## 코스피 10초마다 1번씩 10분동안 갖고오기

```python
time_duration_minute = 10 # 10분
time_sleep = 10

time_duration = time_duration_minute * 60

t0 = time.time()
t_elapsed = time.time() - t0
```

```python
d_l = []

while t_elapsed < time_duration : 
    pyautogui.click()
    
    img = pyautogui.screenshot(region = (pos[0], pos[1], size_x, size_y))
    data = pytesseract.image_to_string(img, lang = 'kor')
    data_s = data[9:17]
    d_l.append(data_s)
    
    print(data_s)
    current_time = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime())
    print(current_time)
    
    t_elapsed = time.time() - t0
    time.sleep(time_sleep)
    pyautogui.click()    
```
