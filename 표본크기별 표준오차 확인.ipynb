```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

plt.style.use('seaborn')
```

## 모집단 크기와 상관없이 표준오차 구하기

```python
p = 0.5

se_l = []
nu_l = np.arange(100, 3100, 100)
for n in nu_l :
    se = 1.96 * np.sqrt(((p*(1-p)) / n))
    se_l.append(se)
plt.plot(nu_l, se_l, marker='o')
```

```python
pd.DataFrame({'표본크기':nu_l, '표준오차':se_l})
```

## 모집단 크기 산정 후 표준오차 구하기

```python
p = 0.5
N = 5000

se_l = []
nu_l = np.arange(100, 3100, 100)
for n in nu_l :
    se = 1.96 * np.sqrt(((p*(1-p)) / n) * ((N-n) / N))
    se_l.append(se)
plt.plot(nu_l, se_l, marker='o')
```

```python
pd.DataFrame({'표본크기':nu_l, '표준오차':se_l})
```
